```python
#Theory Questions:

#1. What is the difference between a function and a method in Python?
'''In Python "function" and "method" both refer to callable objects that perform actions
Functions are used to perform operations that are not necessarily tied to an object or class. They can be called independently and do not require an instance of a class.
'''
#Syntax:
def my_function(param1, param2):
    return param1 + param2
#Calling:
    result = my_function(5, 3)
'''    
Methods are used to define behaviors for objects and classes. They require an instance of the class (for instance methods) or the class itself (for class methods) to be called.'''
#Syntax:
class MyClass:
    def __init__(self, value):
        self.value = value
    
    def my_method(self):
        return self.value * 2
#Calling:
obj = MyClass(10)
result = obj.my_method()
```

```python
#3. What are the different ways to define and call a function in Python?

#1. Basic Function

#Syntax:
def function_name(parameter1, parameter2):
    return result
#Calling:
    result = function_name(arg1, arg2)

#Example:
def add(x, y):
    return x + y

result = add(5, 3)


#2. Default Arguments
#Syntax:
default_value=2
def function_name(parameter1, parameter2=default_value):
    return result

#Calling:
result = function_name(arg1)

#Example:
def greet(name, greeting="Hello"):
    return f"{greeting}, {name}!"

message = greet("Reyansh")

#3. Keyword Arguments
#Syntax:
def function_name(parameter1, parameter2):
    return result

#Calling:
result = function_name(parameter2=value2, parameter1=value1)

#Example:
def describe_pet(pet_name, pet_type):
    return f"{pet_name} is a {pet_type}."

message = describe_pet(pet_type="cat", pet_name="Whiskers")

#4. Variable-Length Arguments

#Positional Variable-Length Arguments (*args):
def sum_all(*numbers):
    return sum(numbers)

total = sum_all(1, 2, 3, 4, 5)

#Keyword Variable-Length Arguments (**kwargs):

def describe_person(**info):
    return info

person_info = describe_person(name="Reyansh", age=30, city="Hyderabad")

#5. Lambda Functions (Anonymous Functions)

square = lambda x: x * x
result = square(4)

#6. Nested Functions

def outer(n):
    def inner(x):
        return x + n
    return inner(5)

result = outer(10)
```

```python
#4. What is the purpose of the `return` statement in a Python function?

'''The return statement in a Python function serves several important purposes:

1. Output a Value from the Function
The primary purpose of the return statement is to send a result back to the caller of the function. This result can be of any data type, such as an integer, string, list, or even another function.
'''

def add(x, y):
    return x + y
'''
2. Terminate the Function Execution
When the return statement is executed, it immediately ends the function’s execution, regardless of whether there are more lines of code left in the function. No code after the return statement will be executed.
'''
def example_function():
    print("This will be printed")
    return
    print("This will not be printed")

example_function()

'''
3. Returning No Value: 
If a function includes a return statement with no value (i.e., just return), it will return None by default. This is similar to a function without a return statement.
'''
def example_function():
    result = example_function()
print(result)

'''
4. Multiple Return Statements
A function can have multiple return statements, often used to handle different conditions or outcomes within the function.
'''
def classify_number(n):
    if n > 0:
        return "Positive"
    elif n < 0:
        return "Negative"
    else:
        return "Zero"

result = classify_number(-5) 
'''
5. Returning Multiple Values
Python allows a function to return multiple values as a tuple. This can be useful for returning more than one result from a function.
'''
def min_max(numbers):
    return min(numbers), max(numbers)

minimum, maximum = min_max([3, 1, 4, 1, 5, 9])
print(minimum)
print(maximum) 
```

```python
#5. What are iterators in Python and how do they differ from iterables?

'''Iterators
An iterator is an object that represents a stream of data. It is an object that follows the iterator protocol.
It includes two main methods:

__iter__(): Returns the iterator object itself. This method is required to make the object an iterator.
__next__(): Returns the next value from the iterator. When there are no more items to return, it raises a StopIteration exception.'''

#example
numbers = [1, 2, 3]
iterator = iter(numbers)
# Iterating through the iterator
print(next(iterator))  # Output: 1
print(next(iterator))  # Output: 2
print(next(iterator))  # Output: 3
# print(next(iterator))  # Raises StopIteration exception

'''
Iterable: An object that can be iterated over (e.g., list, tuple).
Iterator: An object that is used to iterate over an iterable and implements __next__().

Iterable: we can use an iterable directly in a for loop, which automatically gets an iterator from the iterable.
Iterator: we typically use iterators explicitly with next() and StopIteration.'''

# we can also use the iterable directly in a for loop
for number in numbers:
    print(number)
```

```python
#6. Explain the concept of generators in Python and how they are defined.

'''They provide a way to produce items one at a time and only as needed, which can be more memory-efficient than storing all items in memory.
Generators are a powerful feature in Python that allow us to efficiently create iterators with minimal code. They are especially useful for handling large data sets or streams where we don’t want to load all data into memory at once.'''

#Generators are defined using functions with one or more yield statements. 
#syntax:
def generator_function():
    yield value1
    yield value2
#Example    
def square_numbers(n):
    for i in range(n):
        yield i * i

# Create a generator object
squares = square_numbers(5)

# Iterate through the generator
for square in squares:
    print(square)
```

```python
#7. What are the advantages of using generators over regular functions?
'''Generators offer several advantages over regular functions, particularly when it comes to working with large data sets or streams of data. 
Here are some key advantages:'''
'''
Memory Efficiency: 
    Generators are more memory-efficient because they produce items one at a time and do not store the entire sequence in memory.
Lazy Evaluation: 
    Generators produce values on-demand, making them suitable for infinite sequences or large datasets.
Improved Performance: 
    Generators can offer performance benefits by computing values only when needed.
Simplified Code: 
    Generators simplify the creation of iterators, avoiding the need for manual implementation of the iterator protocol.
Handling Streams: 
    Generators are ideal for processing large data streams without consuming excessive memory.'''
```

```python
#8. What is a lambda function in Python and when is it typically used?
'''Lambda functions in Python are useful for creating small, anonymous functions on-the-fly, especially when passing a simple function as an argument to higher-order functions or for short-term use. 
They provide a concise way to write functions but are limited to a single expression and may not be suitable for more complex logic or for scenarios where readability and documentation are important.'''


#Uses
'''
Short-Term Use in Higher-Order Functions:
Lambda functions are often used as arguments to higher-order functions like map(), filter(), and sorted(), where a small function is needed temporarily.

Sorting and Custom Key Functions:
Lambda functions can be used to specify custom sort orders with the sorted() function or other sorting operations.

Simple Operations:
They are useful for defining simple operations in places where a function is needed temporarily and defining a full function would be overkill.

When Defining a Function Inline:
Lambda functions allow you to define a function inline, especially in places where you need to pass a small function as an argument.'''
```

```python
#10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?
'''These functions can often be used together in functional programming styles to process and transform data in powerful and concise ways.

map(): Applies a function to each item in an iterable and returns a new iterable with the transformed values. It is used for transformation.

reduce(): Applies a function cumulatively to items in an iterable to reduce it to a single value. It is used for accumulation or aggregation.

filter(): Applies a function to each item in an iterable to filter out items based on a condition, returning only those items that satisfy the condition. It is used for filtering.'''
```

```python
#4. Write a Python function that checks if a given number is prime or not from 1 to 200

n=int(input("Enter the number"))
def is_prime(num):
    if num <= 1:
        return False
    for i in range(2,num//2+ 1):
        if num % i == 0:
            return False
    return True
if n in range(1,201):
    print(f"{n} is {'prime' if is_prime(n) else 'not prime'}")
```

```python
#5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.

n=int(input("Enter the specified number of terms"))
fib=lambda x:x if x<=1  else fib(x-1)+fib(x-2)
l=[fib(i) for i in range(n)]
print(l)
```

```python
#6. Write a generator function in Python that yields the powers of 2 up to a given exponent.
n=int(input("Enter the exponent"))
def powers_of_2(expo):
    
    for i in range(expo + 1):
        yield 2 ** i
l=[]
for power in powers_of_2(n):
    l.append(power)
print(l)
```

```python
#10. Create a Python program that uses `filter()` to remove all the vowels from a given string

string1 = input("Enter the string")
def not_a_vowel(m):
    l="aeiouAEIOU"
    if m not in l:
        return m
filtered_string = ''.join(filter(not_a_vowel,string1))
print(filtered_string)
```
